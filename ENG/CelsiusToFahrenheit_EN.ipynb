```python
# Import necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Input and output data representing an exact relationship between Celsius and Fahrenheit degrees
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

# Define the neural network layer with one input and one output neuron
layer = tf.keras.layers.Dense(units=1, input_shape=[1])

# Create the sequential model with one layer
model = tf.keras.Sequential([layer])

# Compile the model with optimizer and loss function
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),  # Adam optimizer with learning rate of 0.1 (weights will be updated in one-tenth of the calculated gradient in each training step)
    loss='mean_squared_error'  # Mean squared error loss function
)

# Model training
print('Training starts')
history = model.fit(celsius, fahrenheit, epochs=500, verbose=False)  # Train 500 times
print('Training finished')

# Visualize the loss over epochs
plt.xlabel('Errors')
plt.ylabel('Epochs')
plt.plot(history.history['loss'])
plt.show()
```

```python
print('Testing the program')
result = model.predict([100.0])
print('The result of 100 Celsius is '+str(result)+' Fahrenheit')
```
