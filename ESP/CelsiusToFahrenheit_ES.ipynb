```python
# Importar las bibliotecas necesarias
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada y salida que representan una relacion Exacata entre grados Celsius y Fahrenheit
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

# Definir la capa de la red neuronal con una neurona de salida y de entrada
capa = tf.keras.layers.Dense(units=1, input_shape=[1])

# Crear el modelo secuencial con una capa
modelo = tf.keras.Sequential([capa])

# Compilar el modelo con el optimizador y la función de pérdida
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),  # Optimizador Adam con tasa de aprendizaje de 0.1(los pesos se actualizarán en una décima parte del gradiente calculado en cada paso de entrenamiento.)
    loss='mean_squared_error'  # Función de pérdida de error
)

# Entrenamiento del modelo
print('Comenzamos entrenamiento')
historial = modelo.fit(celsius, fahrenheit, epochs=500, verbose=False)  # Entrenar 500 veces
print('Entrenamiento terminado')

# Visualización de la pérdida a lo largo de las épocas
plt.xlabel('Errors')
plt.ylabel('Epochs')
plt.plot(historial.history['loss'])
plt.show()
```

```python
print('Pongamos a prueba el programa')
resultado= modelo.predict([100.0])
print('El resultado de 100 celsius son '+str(resultado)+' Fahrenheit')
```
