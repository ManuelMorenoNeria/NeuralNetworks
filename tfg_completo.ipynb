```python
import os
import cv2  # Para el procesamiento de imágenes
import numpy as np  # Importa NumPy para operaciones matemáticas y manejo de matrices
from sklearn.model_selection import train_test_split  # Importa train_test_split para dividir datos
import tensorflow as tf  # Importa TensorFlow para el aprendizaje profundo
from tensorflow.keras import layers, models  # Importa Keras para la construcción de modelos

# Función para cargar imágenes desde un directorio dado y redimensionarlas
def cargar_imagenes(directory, tamano=(28, 28)):
    imagenes = []  # Lista para almacenar las imágenes de letras
    etiquetas = []  # Lista para almacenar las etiquetas (nombres de impresores)

    for impresor in os.listdir(directory):
        impresor_directory = os.path.join(directory, impresor)  # Ruta del directorio del impresor actual (IMPRESOR1)
        print("ESTAMOS EN",impresor_directory)
        # Verifica si el directorio es una carpeta
        if os.path.isdir(impresor_directory):
            # Itera sobre los subdirectorios de cada impresor
            for subdirectory in os.listdir(impresor_directory):
                subdirectory_path = os.path.join(impresor_directory, subdirectory)  # Ruta del subdirectorio (E y X)
                print("TAMO EN",subdirectory_path)
                # Verifica si el subdirectorio es una carpeta
                if os.path.isdir(subdirectory_path):
                    for archivo in os.listdir(subdirectory_path):
                        imagen_path = os.path.join(subdirectory_path, archivo)  # Ruta de la imagen actual
                        print("Intentando cargar:", imagen_path)
                        img = cv2.imread(imagen_path, cv2.IMREAD_GRAYSCALE)  # Lee la imagen en escala de grises

                        if img is not None:
                            print("Imagen cargada con éxito.")
                            imagen_redim = cv2.resize(img, tamano)  # Redimensiona la imagen al tamaño objetivo
                            imagenes.append(imagen_redim)  # Agrega la imagen redimensionada a la lista de imágenes
                            etiquetas.append(impresor)  # Agrega la etiqueta del impresor a la lista de etiquetas
                        else:
                          if os.path.isdir(imagen_path):
                            for archivos in os.listdir(imagen_path):
                              imagen_path2=os.path.join(imagen_path,archivos)
                              print("Intentando cargar:", imagen_path2)
                              img = cv2.imread(imagen_path2, cv2.IMREAD_GRAYSCALE)

                              if img is not None:
                                  print("Imagen cargada con éxito.")
                                  imagen_redim = cv2.resize(img, tamano)  # Redimensiona la imagen al tamaño objetivo
                                  imagenes.append(imagen_redim)  # Agrega la imagen redimensionada a la lista de imágenes
                                  etiquetas.append(impresor)  # Agrega la etiqueta del impresor a la lista de etiquetas
                          else:
                            print("Error al cargar la imagen:", imagen_path2)  # Mensaje de error que nos avise si algo no se pudo cargar bien
                else:
                   img = cv2.imread(subdirectory_path, cv2.IMREAD_GRAYSCALE)  # Lee la imagen en escala de grises
                   if img is not None:
                     print("Imagen cargada con éxito.")
                     imagen_redim = cv2.resize(img, tamano)  # Redimensiona la imagen al tamaño objetivo
                     imagenes.append(imagen_redim)  # Agrega la imagen redimensionada a la lista de imágenes
                     etiquetas.append(impresor)  # Agrega la etiqueta del impresor a la lista de etiquetas



        else:
          for foto in os.listdir(impresor_directory):
            path_foto= os.path.join(impresor_directory,foto)
            print("Intentando cargar:", path_foto)
            img = cv2.imread(path_foto, cv2.IMREAD_GRAYSCALE)  # Lee la imagen en escala de grises
            if img is not None:
              print("Imagen cargada con éxito.")
              imagen_redim = cv2.resize(img, tamano)  # Redimensiona la imagen al tamaño objetivo
              imagenes.append(imagen_redim)  # Agrega la imagen redimensionada a la lista de imágenes
              etiquetas.append(impresor)  # Agrega la etiqueta del impresor a la lista de etiquetas

    # Convierte las listas de imágenes y etiquetas en matrices
    imagenes = np.array(imagenes)
    etiquetas = np.array(etiquetas)

    return imagenes, etiquetas


# Carga las imágenes y etiquetas
imagenes, etiquetas = cargar_imagenes("/content/LETRAS2/impresor1")

# Divide los datos en conjuntos de entrenamiento y prueba
train_images, test_images, train_labels, test_labels = train_test_split(imagenes, etiquetas)

# Convertir etiquetas a valores numéricos
train_labels_numeric = np.where(train_labels == 'E', 0, 1)
test_labels_numeric = np.where(test_labels == 'E', 0, 1)

# Convertimos los valores de píxeles de las imágenes en valores entre 0 y 1
train_images = train_images / 255.0
test_images = test_images / 255.0

# Define lel modelo CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)
])

# Compilo el modelo
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Entreno el modelo
model.fit(train_images, train_labels_numeric, epochs=10, validation_data=(test_images, test_labels_numeric))

# Evaluo el modelo y saco por pantalla la precisión del modelo
test_loss, test_acc = model.evaluate(test_images, test_labels_numeric, verbose=2)
print('Porcentaje de acierto:', test_acc*100,'%')
```
